```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
import numpy as np
import seaborn as sns

from sklearn.datasets import fetch_openml
```

```python
df = pd.read_csv('sale_data.csv')
df['date'] = pd.to_datetime(df['date'])
#df['date'] = df['date'].dt.strftime('%Y/%m/%d')
df
```

```python
lt = pd.DataFrame({'date1': pd.to_datetime(['2008-11-29'])})
```

```python
df['recency'] = (lt['date1'][0] - df['date']).dt.days
df
```

```python
df['monetary'] = df['amount'] * df['single_price']
df['frequency'] = 1
df
```

```python
pivot_3 = np.round(pd.pivot_table(df, values='monetary', 
                                index='customer_id', 
                                aggfunc=np.sum),2)
pivot_3
```

```python
pivot_2 = np.round(pd.pivot_table(df, values='recency', 
                                index='customer_id', 
                                aggfunc=np.min),2)
pivot_2
```

```python
pivot_1 = np.round(pd.pivot_table(df, values='frequency', 
                                index='customer_id', 
                                aggfunc=np.sum),2)
pivot_1
```

```python
df_rfm = pd.DataFrame(pivot_1)
df_rfm = pd.merge(df_rfm, pivot_2, left_index=True, right_index=True)
df_rfm = pd.merge(df_rfm, pivot_3, left_index=True, right_index=True)
df_rfm
```

```python
#Normalize
for column in df_rfm.columns: 
    df_rfm[column] = df_rfm[column]  / df_rfm[column].abs().max() 
      
display(df_rfm) 
```

```python
df_rfm['frequency_score'] = round(df_rfm['frequency']*5,0)
df_rfm['recency_score'] = round(df_rfm['recency']*5,0)
df_rfm['monetary_score'] = round(df_rfm['monetary']*5,0)
df_rfm
```

```python
RFM = df_rfm
RFM['RFM_Score'] = (df_rfm['recency_score']*100) + (df_rfm['monetary_score']) + (df_rfm['frequency_score']*10)
RFM['RF_Score'] = ((df_rfm['recency_score']*10) + (df_rfm['frequency_score'])).astype(str).apply(lambda x: x.replace('.0',''))
RFM
RFM
```

```python
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions',
    r'[1-5]': 'churn'
}
RFM['Segment'] = RFM['RF_Score'].replace(seg_map, regex=True)
RFM = RFM.sort_values(['Segment'], ascending=[False])
RFM
```
